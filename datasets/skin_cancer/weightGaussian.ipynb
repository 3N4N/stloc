```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


tsv_file = open("GSE144239_ST_P2_S1_counts.tsv")
dataset= pd.read_csv(tsv_file, delimiter="\t")
rowNames = dataset.index.values
temp = rowNames[0].split("x")
print(temp)


xVal=[]
yVal=[]
w=[]
for i in rowNames :
    temp = i.split("x")
    xVal.append(int(temp[0]))
    yVal.append(int(temp[1]))
    
mu=[30,32]
l=0.01

for  i in range(len(xVal)):
    dist= ((xVal[i] -mu[0])**2 +( yVal[i] -mu[1])**2)/ l**2
    w.append( np.exp( - dist))
    
plt.scatter(xVal,yVal,c=w)
```
